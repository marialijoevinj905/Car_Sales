```python
# Function for log transformation of the column
def log_transform(d,col):
    for colname in col:
        if (d[colname] == 1.0).all():
            d[colname + '_log'] = np.log(d[colname]+1)
        else:
            d[colname + '_log'] = np.log(d[colname])
    d.info()
```

```python
d=df.copy()
log_transform(d,['Kilometers_Driven','Price'])
```

```python
#Log transformation of the feature 'Kilometers_Driven'
sns.distplot(d["Kilometers_Driven_log"], axlabel="Kilometers_Driven_log");
```

```python
#Bivariate
fig, axarr = plt.subplots(4, 2, figsize=(12, 18))
d.groupby('Location')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][0], fontsize=12,color='purple')
axarr[0][0].set_title("Location Vs Price", fontsize=18)
d.groupby('Transmission')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][1], fontsize=12,color='darkgreen')
axarr[0][1].set_title("Transmission Vs Price", fontsize=18)
d.groupby('Fuel_Type')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][0], fontsize=12,color='deeppink')
axarr[1][0].set_title("Fuel_Type Vs Price", fontsize=18)
d.groupby('Owner_Type')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][1], fontsize=12,color='saddlebrown')
axarr[1][1].set_title("Owner_Type Vs Price", fontsize=18)
d.groupby('Brand')['Price_log'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][0], fontsize=12,color='indigo')
axarr[2][0].set_title("Brand Vs Price", fontsize=18)
d.groupby('Model')['Price_log'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][1], fontsize=12,color='grey')
axarr[2][1].set_title("Model Vs Price", fontsize=18)
d.groupby('Seats')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][0], fontsize=12,color='tan')
axarr[3][0].set_title("Seats Vs Price", fontsize=18)
d.groupby('Car_Age')['Price_log'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][1], fontsize=12,color='aqua')
axarr[3][1].set_title("Car_Age Vs Price", fontsize=18)
plt.subplots_adjust(hspace=1.0)
plt.subplots_adjust(wspace=.5)
sns.despine()
```
